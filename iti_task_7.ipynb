```python
# TASK 2

A=[85,90,88,92,78,95,89,91,82,87]
B=[78,82,80,85,76,88,84,89,83,79]

import numpy as np
from scipy import stats

t_test = stats.ttest_ind(A, B, equal_var=True)

t_stat = t_test.statistic

df = t_test.df

p_value = t_test.pvalue

if p_value < 0.05:
  print("there is significant difference between the means of the two groups.")
else:
  print("there is not significant difference between the means of the two groups.")
```
