```python
class ExhaustiveLibrary:
    
    def DFS(graph):
        start_node = list(graph.keys())[0]  
        visited = []
        stack = [start_node]
        visited.append(start_node)
        while stack:
            node = stack.pop()
            print(node, end=" ")
            for neighbour in graph[node]:
                if neighbour not in visited:
                    visited.append(neighbour)
                    stack.append(neighbour)

   
    def BFS(graph):
        start_node = list(graph.keys())[0]  
        visited = []
        queue = [start_node]
        visited.append(start_node)
        while queue:
            node = queue.pop(0)
            print(node, end=" ")
            for neighbour in graph[node]:
                if neighbour not in visited:
                    visited.append(neighbour)
                    queue.append(neighbour)

    
    def BID(graph):
        start_node = list(graph.keys())[0] 
        goal_node = list(graph.keys())[-1]  
        visited_forward = []
        visited_backward = []
        queue_forward = [start_node]
        queue_backward = [goal_node]
        visited_forward.append(start_node)
        visited_backward.append(goal_node)
        while queue_forward and queue_backward:
            
            forward_node = queue_forward.pop(0)
            print(forward_node, end=" ")
            for neighbour in graph[forward_node]:
                if neighbour not in visited_forward:
                    visited_forward.append(neighbour)
                    queue_forward.append(neighbour)

            
            backward_node = queue_backward.pop(0)
            print(backward_node, end=" ")
            for neighbour in graph[backward_node]:
                if neighbour not in visited_backward:
                    visited_backward.append(neighbour)
                    queue_backward.append(neighbour)

            
            if set(visited_forward).intersection(visited_backward):
                print("Goal node found!")
                return




graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("DFS:")
ExhaustiveLibrary.DFS(graph)
print("\nBFS:")
ExhaustiveLibrary.BFS(graph)
print("\nBID:")
ExhaustiveLibrary.BID(graph)
```

```python
import os

def reverse_directory_bfs(directory):
    queue = [directory]
    visited = set()

    while queue:
        current_dir = queue.pop(0)
        visited.add(current_dir)

        
        items = os.listdir(current_dir)

        for item in items:
            item_path = os.path.join(current_dir, item)

            if os.path.isdir(item_path) and item_path not in visited:
                queue.append(item_path)

        
        items.reverse()

        for item in items:
            item_path = os.path.join(current_dir, item)
            
            if os.path.isdir(item_path) and item_path not in visited:
                queue.append(item_path)

        
        items.reverse()

def reverse_directory_dfs(directory):
    stack = [directory]
    visited = set()

    while stack:
        current_dir = stack.pop()
        visited.add(current_dir)

        
        items = os.listdir(current_dir)

        for item in items:
            item_path = os.path.join(current_dir, item)

            if os.path.isdir(item_path) and item_path not in visited:
                stack.append(item_path)

        
        items.reverse()

        for item in items:
            item_path = os.path.join(current_dir, item)
            
            if os.path.isdir(item_path) and item_path not in visited:
                stack.append(item_path)

        
        items.reverse()


directory_path =  {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
reverse_directory_bfs(directory_path)
reverse_directory_dfs(directory_path)
```
